# Fit after cleaning data

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
%matplotlib inline

mean_flux = [5614.533804359948, 5095.131427130742, 4165.479808934043, 3075.1703653426575, 1944.662135516696, 1378.8958518315217, 1163.1671899370629]
errs = [4.553740187292434, 12.22975323811172, 3.8222262302571135, 5.95941706605791, 3.8409157150887308, 3.4741371890614636, 2.923110102944669]
angles = [0, 15, 30, 45, 60, 75, 90]
rad_angles = [a * (np.pi / 180) for a in angles]
plt.errorbar(angles, mean_flux,  yerr=errs, fmt='o', capsize=4)

def objective(flux, A, B, n):
    return A + B * (np.cos(flux)**(n)) 

popt, var = curve_fit(objective, rad_angles, mean_flux, p0=[1208, 4266, 2], sigma = errs, maxfev = 8000)


A, B, n = popt
print('A: %.5f B: %.5f n: %.5f' % (A, B, n))

#covariance matrix
#print(var)

#one std errors of parameters
perr = np.sqrt(np.diag(var))
print('std of n: ', perr[2])

d = (n - 2) / 2
print('percent error: ', d*100)

fit = objective(rad_angles, A, B, n)
ideal = objective(rad_angles, A, B, 2)

x = np.arange(0, 90, 0.1)
y_ideal = A + B * (np.cos(x * np.pi/180)**2)
y_fit = A + B * (np.cos(x * np.pi/180)**n)

#plt.plot(angles, fit, linestyle='dashed')
#plt.plot(angles, ideal, c='r')
plt.plot(x, y_fit, linestyle='dashed')
plt.plot(x,y_ideal, c='r')

plt.xlabel('Angle (deg)')
plt.ylabel('Flux (events/m$^2$/min)')
plt.show()
```
